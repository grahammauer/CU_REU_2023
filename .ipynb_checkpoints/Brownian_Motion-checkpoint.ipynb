```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to create a Gaussian Orthoganal Ensemble
def Gaussian_Orthogonal_Ensemble(n, **kwargs):
    
    '''
    Generates a nxn Gaussian Orthogonal Ensemble
    '''
    
    # Define a normally distributed array
    H = np.random.normal(0, 1, size=(n,n))
    
    # Make it symmetric
    H = np.triu(H) + np.triu(H, k=1).T
    
    # Multiply the diagonal values by a factor of sqrt(2)
    np.fill_diagonal(H, np.sqrt(2) * np.diag(H))
    
    return H

# Define a function to create a Gaussian Unitary Ensemble
def Gaussian_Unitary_Ensemble(n, **kwargs):
    
    '''
    Generates a nxn Gaussian Unitary Ensemble
    '''
    
    # Define a normally distributed array with imaginary component
    H = np.random.normal(0, 1, size=(n,n)) + 1j * np.random.normal(0, 1, size=(n,n))
    
    # Make it symmetric
    H = np.triu(H) + np.conjugate(np.triu(H, k=1).T)
    
    # Multiply the diagonal values by a factor of sqrt(2)
    np.fill_diagonal(H, np.sqrt(2) * np.real(np.diag(H)))
    
    return H

def eigenvalues(n, **kwargs):
    
    '''
    Computes the eigenvalues for a nxn Gaussian Orthogonal Ensemble
    '''
    
    eigs = np.linalg.eig(Gaussian_Orthogonal_Ensemble(n))[0]
    
    return eigs

def eigenvalues_GUE(n, **kwargs):
    
    '''
    Computes the eigenvalues for a nxn Gaussian Unitary Ensemble
    '''
    
    eigs = np.real(np.linalg.eig(Gaussian_Unitary_Ensemble(n))[0])
    
    return eigs

def sqrt_interp(t, p1, p2, **kwargs):
    
    # Create a copy of each point
    pt1 = p1.copy()
    pt2 = p2.copy()
    
    # Project point 1 to x axis
    pt1[1] = 0
    pt2[1] = p2[1] - p1[1]
    
    # Calculate positive and reflect over y axis if needed
    if pt2[1] - pt1[1] < 0:
        
        sqrt_interp = -1 * np.sqrt(np.interp(t, [pt1[0], pt2[0]], [pt1[1]**2, pt2[1]**2]))
    
    else:
        
        sqrt_interp = np.sqrt(np.interp(t, [pt1[0], pt2[0]], [pt1[1]**2, pt2[1]**2]))
    
    # Undo projection
    sqrt_interp += p1[1]
    
    # Reflect and translate
    sqrt_interp = -1 * sqrt_interp + p1[1] + p2[1]
    
    return sqrt_interp

# sqrt_interp() test
point1 = [1,-10]
point2 = [2,5]

t = np.linspace(point1[0], point2[0], 1000)
#plt.figure(figsize=(7,7))
#plt.plot(np.flip(t), sqrt_interp(t, point1, point2))
#plt.scatter(point1[0], point1[1], color='tab:orange')
#plt.scatter(point2[0], point2[1], color='tab:orange')
#plt.title('Square Root Interpolation Test')
#plt.show()

def n_Brownian_Motions_GOE(n, n_i, T, steps, dt):
    
    '''
    Simulates the Brownian Motion of n particles with starting positions of eigenvalues of a GOE.
    '''
    
    # Create an array of positions +  assign initial values
    positions = np.zeros((int(np.ceil((T+dt) / dt)), n))
    positions[0] = n_i
    
    # For each timestep after the initial
    for i in range(len(positions) - 1):
        
        # Generate some eigenvalues for the motion
        eigs = np.sort(eigenvalues(n))
        
        # Calculate the Coulombic interation
        for j in range(n):
            
            coulombic = 0
            
            for k in range(n):
            
                if k != j:
                
                    coulombic += 1 / (positions[i, j] - positions[i, k])
            
            # Assign that interaction to each particle and calculate the path for that timestep
            positions[i+1, j] = positions[i, j] + np.sqrt(2 * dt / n) * eigs[j] + coulombic / n
    
    # Slice out the wanted values
    positions_interp = positions[::int(T / dt) // steps]
        
    return positions, positions_interp

def n_Brownian_Motions_GUE(n, n_i, T, steps, dt):
    
    '''
    Simulates the Brownian Motion of n particles with starting positions of eigenvalues of a GUE.
    '''
    
    # Create an array of positions +  assign initial values
    positions = np.zeros((int(np.ceil((T+dt) / dt)), n))
    positions[0] = n_i
    
    # For each timestep after the initial
    for i in range(len(positions) - 1):
        
        # Generate some eigenvalues for the motion
        eigs = np.sort(eigenvalues_GUE(n))
        
        # Calculate the Coulombic interation
        for j in range(n):
            
            coulombic = 0
            
            for k in range(n):
            
                if k != j:
                
                    coulombic += 1 / (positions[i, j] - positions[i, k])
            
            # Assign that interaction to each particle and calculate the path for that timestep
            positions[i+1, j] = positions[i, j] + np.sqrt(2 * dt / n) * eigs[j] + coulombic / n
    
    # Slice out the wanted values
    positions_interp = positions[::int(T / dt) // steps]
        
    return positions, positions_interp

############################
# Brownian Motion Conditions
############################
n = 5
n_i = np.sort(eigenvalues(n))
T = 1
steps = 5 # Stepsize for sqrt interpolation
dt = 0.02
interp_mesh_size = 1000

##########
# Plotting
##########
positions, positions_interp = n_Brownian_Motions_GOE(n, n_i, T, steps, dt)
t = np.linspace(0,T,int(np.ceil((T+dt) / dt)))

plt.figure(figsize=(10,10))
plt.plot(t, positions)
plt.title(f'Brownian Motion without Square Root Interpolation')
plt.show()

plt.figure(figsize=(10,10))
plt.plot(t, positions)
plt.title(f'Brownian Motion with Square Root Interpolation every {steps} steps')

for i in range(n):
    
    for j in range((int(T / dt) // steps)):
        
        # Calculate where each 'point' is [t, y], same as in the example above.
        point1 = [T - (j + 1) * dt * steps, positions[-1 * (j+1) * steps-1, i]]
        point2 = [T - j * dt * steps, positions[-1 * j * steps-1, i]]

        t = np.linspace(point1[0], point2[0], interp_mesh_size)
        plt.plot(np.flip(t), sqrt_interp(t, point1, point2), color='tab:red', lw=0.8)
            
plt.show()

############################
# Brownian Motion Conditions
############################
n = 5
n_i = np.sort(eigenvalues(n))
T = 1
steps = 5 # Stepsize for sqrt interpolation
dt = 0.02
interp_mesh_size = 1000

##########
# Plotting
##########
positions, positions_interp = n_Brownian_Motions_GUE(n, n_i, T, steps, dt)
t = np.linspace(0,T,int(np.ceil((T+dt) / dt)))

plt.figure(figsize=(10,10))
plt.plot(t, positions)
plt.title(f'Brownian Motion without Square Root Interpolation')
plt.show()

plt.figure(figsize=(10,10))
plt.plot(t, positions)
plt.title(f'Brownian Motion with Square Root Interpolation every {steps} steps')

for i in range(n):
    
    for j in range((int(T / dt) // steps)):
        
        # Calculate where each 'point' is [t, y], same as in the example above.
        point1 = [T - (j + 1) * dt * steps, positions[-1 * (j+1) * steps-1, i]]
        point2 = [T - j * dt * steps, positions[-1 * j * steps-1, i]]

        t = np.linspace(point1[0], point2[0], interp_mesh_size)
        plt.plot(np.flip(t), sqrt_interp(t, point1, point2), color='tab:red', lw=0.8)
            
plt.show()
    
```

# Putting these results into the code as drivers

```python
# Pulling out each path as a driver
interp_functions = np.zeros((n, (int(T / dt) // steps) * interp_mesh_size))

for i in range(n):
    
    for j in range((int(T / dt) // steps)):
        
        point1 = [T - (j + 1) * dt * steps, positions[-1 * (j+1) * steps-1, i]]
        point2 = [T - j * dt * steps, positions[-1 * j * steps-1, i]]

        t = np.flip(np.linspace(point1[0], point2[0], interp_mesh_size))
        interp_functions[i, (int(T/dt) // steps - j - 1) * interp_mesh_size:(int(T/dt) // steps - j) * interp_mesh_size] = sqrt_interp(t, point1, point2)
        
# Plot it to make sure it looks correct
plt.figure(figsize=(7,7))
for i in range(n):
    
    t = np.linspace(0,T,(int(T / dt) // steps) * interp_mesh_size)
    plt.plot(t, interp_functions[i])
    
plt.show()
```

```python
from scipy.integrate import solve_ivp

C = 1

# Derivative definition
def complex_function(t,y):
    
    x = y[0]
    y = y[1]
    
    dxdt = 0
    dydt = 0
    
    for i in range(n):
        
        func = np.interp(T-t, np.linspace(0,T,(int(T / dt) // steps) * interp_mesh_size), interp_functions[i])
        
        dxdt += -2 * (x - C * func) / ( (x - C * func) ** 2 + y ** 2 )
        
        dydt += 2 * y / ( (x - C * func) ** 2 + y ** 2 )

    dxdt /= 2
    
    dydt /= 2

    return np.array([dxdt, dydt])

t_max = 2

x_points = np.linspace(-40,40,20)
y_points = np.linspace(1, 5, 20)

x_sols = []
y_sols = []
t_sols = []

print('Starting Sims')

for x0 in x_points:
    
    print(x0)
    
    for y0 in y_points:
        
        t_span = [0,t_max]
        
        z0 = np.array([x0,y0])
        
        sol = solve_ivp(complex_function, t_span, z0, method='BDF', atol=1e-15, rtol=1e-12)

        x = sol.y[0]
        y = sol.y[1]
        t = sol.t
        
        x_sols.append(x)
        y_sols.append(y)
        t_sols.append(t)  
        
print('Plotting')

plt.figure(figsize=(7,7))

for i in range(len(x_points) * len(y_points)):

        plt.plot(x_sols[i], y_sols[i], color='tab:blue')

plt.xlabel('Reals')
plt.ylabel('Imaginary')
plt.show()
```
