# Checking Analytical Solutions

```python
# Vectorized array operations
import numpy as np

# Plotting
import matplotlib.pyplot as plt

# Blackbox ODE solver
from scipy.integrate import solve_ivp

# Constant
C=0
h0 = [2]

# Derivative definition
def positive_numerator_no_C(t,h):
    
    dhdt = 2 / h[-1]
    
    return [dhdt]

# Analytical Solution
def analytical_solution(t, C=0, h_0 = h0[0]):
    
    return np.sqrt(4*t + (C + h_0)**2) - C

# Maximum runtime, if no singularity increase this. No penalty if this is larger than 
t_max = 10
t_span = [0,t_max]

# Run the ODE solver and pull out solutions
sol = solve_ivp(positive_numerator_no_C, t_span, h0, method='BDF', atol=1e-15, rtol=1e-12)
h = sol.y[0]
t = sol.t

# Plot numerical and analytical solutions
plt.figure(figsize=(5,5), dpi=200)
plt.title(f'positive_numerator_no_C()\nc={C}, h_0={h0[0]}')
plt.plot(t,h, label='BDF')
t_fine=np.linspace(0,t_max,100000)
plt.plot(t_fine,analytical_solution(t_fine), label='Analytical Solution', color='tab:orange', linestyle='--')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.legend()
plt.show()

# Plot Error
plt.figure(figsize=(5,5), dpi=200)
plt.title(f'positive_numerator_no_C()\nAbsolute Error')
plt.plot(t,analytical_solution(t) - h, label='Error', color='tab:orange')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.legend()
plt.savefig('BDF.png')
plt.show()

# Plot timesteps
plt.figure(figsize=(5,5), dpi=200)
plt.title('positive_numerator_no_C()\nTimestep')
plt.plot(t[1:],t[1:] - t[:-1], label=f'BDF {len(t)} steps')
plt.legend()
plt.yscale('log')
plt.xlabel('t')
plt.ylabel('Timestep')
plt.show()
```

# Complex Domain

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

C = 1
# Lambda definition
def lambda_(C,t):
    
    return C * np.sqrt(t)

# Derivative definition
def complex_function(t,y):
    
    x = y[0]
    y = y[1]

    dxdt = 0.5 * ( -2 * (x - lambda_(C, t)) / ( (x - lambda_(C, t)) ** 2 + y ** 2 ) + -2 * (x + lambda_(C, t)) / ( (x + lambda_(C, t)) ** 2 + y ** 2 ) )
    
    dydt = 0.5 * ( 2 * y / ( (x - lambda_(C, t)) ** 2 + y ** 2 ) + 2 * y / ( (x + lambda_(C, t)) ** 2 + y ** 2 ) )

    return np.array([dxdt, dydt])

t_max = 10

x_points = np.linspace(-10, 10, 20)
y_points = np.linspace(0, 10, 20)

x_sols = []
y_sols = []
t_sols = []

for x0 in x_points:
    
    for y0 in y_points:
        
        t_span = [0,t_max]
        
        z0 = np.array([x0,y0])
        
        sol = solve_ivp(complex_function, t_span, z0, method='BDF', atol=1e-15, rtol=1e-12)

        x = sol.y[0]
        y = sol.y[1]
        t = sol.t
        
        x_sols.append(x)
        y_sols.append(y)
        t_sols.append(t)    

for i in range(len(x_points) * len(y_points)):

        plt.plot(x_sols[i], y_sols[i], color='tab:blue')

plt.show()

anim_times = np.linspace(0,t_max, 121)

# Pull the animation data at the desired times and assign it to an array
grid = np.zeros((len(anim_times), len(x_points), len(y_points)))

for i in range(len(x_points) * len(y_points)):
    
    grid[:,int(np.floor(i / len(y_points))), int(np.mod(i, len(y_points)))] = np.interp(anim_times, t_sols[i], x_sols[i])

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
import numpy as np

# Create a function to generate your plots
def generate_plot(frame):
    
    #print(frame)

    # Clear the current axes
    plt.cla()

    # Plot the data
    for i in range(len(x_points) * len(y_points)):
        
        plt.scatter(np.interp(anim_times[frame], t_sols[i], x_sols[i]),np.interp(anim_times[frame], t_sols[i], y_sols[i]), color='tab:blue', s=0.2)

    # Set plot title and labels, if desired
    plt.title(f"t={anim_times[frame]:0.2f}")

    # Return the plot or figure object
    return plt

# Set up the figure and axes
fig, ax = plt.subplots()

# Create the animation
anim = animation.FuncAnimation(fig, generate_plot, frames=len(anim_times), interval=200)

# Set up the writer
writer = PillowWriter(fps=10)

# Save the animation as an MP4 file
anim.save('animation_new.gif', writer=writer)
plt.close()



```
